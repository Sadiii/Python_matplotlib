#### Filling Area with Line Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv('dataset.csv')
ages = data["Age"]
all_devs = data['All_Devs']
py_dev = data['Python']
js_dev = data['JavaScript']

plt.plot(ages, all_devs, label = "All_Devs")

plt.plot(ages, py_dev,  label = "Python_Dev")

overall_median = 57287
#plt.fill_between(ages, py_dev,overall_median, alpha = 0.25)
plt.fill_between(ages, py_dev, overall_median, where = (py_dev > overall_median), color = 'green', interpolate=True, alpha = 0.25)
plt.fill_between(ages, py_dev, overall_median, where = (py_dev <= overall_median), color = 'red', interpolate=True, alpha = 0.25)

plt.title("Developer w.r.t Ages")
plt.legend()
plt.grid(True)
plt.xlabel("Ages")
plt.ylabel('Salaries')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv('dataset.csv')
ages = data["Age"]
all_devs = data['All_Devs']
py_dev = data['Python']
js_dev = data['JavaScript']

plt.plot(ages, all_devs, label = "All_Devs", linestyle = "--", color = "#444444")

plt.plot(ages, py_dev,  label = "Python_Dev", color = '#5a7d9a')

overall_median = 57287
#plt.fill_between(ages, py_dev,overall_median, alpha = 0.25)
plt.fill_between(ages, py_dev, all_devs, where = (py_dev > all_devs),
                 color = 'green', interpolate=True, alpha = 0.25, label = 'Above Average')
plt.fill_between(ages, py_dev, all_devs, where = (py_dev <= all_devs), 
                 color = 'red', interpolate=True, alpha = 0.25, label = "Below Average")

plt.title("Developer w.r.t Ages")
plt.legend()
plt.grid(True)
plt.xlabel("Ages")
plt.ylabel('Salaries')

plt.show()
```
