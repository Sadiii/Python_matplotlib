```python
import csv
import matplotlib.pyplot as plt
from collections import Counter
plt.style.use('fivethirtyeight')

with open('data.csv') as file:
    csv_reader = csv.DictReader(file)
    #print(next(csv_reader))
    #row = next(csv_reader)
    #print(row['LanguagesWorkedWith'])
    file_counter = Counter()
    for row in csv_reader:
        file_counter.update(row['LanguagesWorkedWith'].split(";"))

#print(file_counter.most_common(15))
languages = [x[0] for x in file_counter.most_common(15)] 
popularity = [x[1] for x in file_counter.most_common(15)] 

languages.reverse()
popularity.reverse()
#print(languages)
#print(popularity)

plt.barh(languages, popularity)
plt.title("Most Popular Languages")
#plt.xticks(ticks = indexes, labels = ages)
#plt.ylabel("Languages")
plt.xlabel("Number of Users")


plt.show()
```

```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
from collections import Counter

plt.style.use('fivethirtyeight')

data = pd.read_csv('data.csv')
ids = data['Responder_id']
lang_data = data['LanguagesWorkedWith']

file_counter = Counter()

for row in lang_data:
    file_counter.update(row.split(';'))

languages = [x[0] for x in file_counter.most_common(15)]
popularity = [x[1] for x in file_counter.most_common(15)]
languages.reverse()
popularity.reverse()


plt.barh(languages, popularity)
plt.title("Most Popular Languages")
#plt.xticks(ticks = indexes, labels = ages)
#plt.ylabel("Languages")
plt.xlabel("Number of Users")


plt.show()
```
